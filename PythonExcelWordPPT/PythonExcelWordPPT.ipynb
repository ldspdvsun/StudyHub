### 读写工作表

```python
# 导入xlwings库,并且取别名为xw
import xlwings as xw

# 调用xw库的App类与Excel建立连接
app = xw.App(visible=True, add_book=False)

# 新建一个工作簿
workbook = app.books.add()
# 输出工作簿名字
print(workbook.name)

# 选中工作表sheet1
worksheet = workbook.sheets['sheet1']
# 输出工作表名字
print(worksheet.name)

# 把“地区”写入单元格A1中
worksheet.range('A1').value = '地区'
# 读取单元格A1的内容
value = worksheet.range('A1').value
# 输出A1的内容
print(value)

# 获取A1的行高
row_height = worksheet.range('A1').row_height
# 获取A1的列宽
column_width = worksheet.range('A1').column_width
print('行高是：{}，列宽是：{}'.format(row_height,column_width))

```
