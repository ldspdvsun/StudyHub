```python
# coding:utf-8
# @File    : docx1.py
# @Time    : 2022/6/29 22:08
# @Author  : Sun
# docx word文档的处理
import xlrd
import xlrd2
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT, WD_LINE_SPACING
from docx import Document

# 在doc开头调用进行设置
from docx.oxml.ns import qn
from docx.shared import Pt, Cm, Mm, RGBColor


# 设置全局文本样式变量
def docxinitial(doc):
    # 设置正文字体类型、大小
    doc.styles["Normal"].font.name = u'宋体'
    doc.styles["Normal"].font.size = Pt(12)
    doc.styles["Normal"]._element.rPr.rFonts.set(qn('w:eastAsia'), u'宋体')
    doc.styles["Normal"]._element.line_spacing = Pt(20)
    # 设置首行缩进2字符
    # doc.styles["Normal"].paragraph_format.first_line_indent = Pt(20) * 1
    # 设置页眉
    header = doc.sections[0].header
    pheader = header.paragraphs[0]  # 获取页眉的第一个段落
    ph_header = pheader.add_run('操作系统测试说明')
    ph_header.font.name = u'宋体'  # 设置页眉字体样式
    ph_header._element.rPr.rFonts.set(qn('w:eastAsia'), u'宋体')
    ph_header.font.size = Pt(10.5)  # 设置页眉字体大小
    # ph_header.bold = True  # 页眉字体加粗
    pheader.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER  # 页眉对齐方式设为居中
    # 设置页脚
    footer = doc.sections[0].footer
    pfooter = footer.paragraphs[0]  # 获取页脚的第一个段落
    ph_footer = pfooter.add_run('XX公司')
    ph_footer.font.name = u'宋体'  # 设置页眉字体样式
    ph_footer._element.rPr.rFonts.set(qn('w:eastAsia'), u'宋体')
    ph_footer.font.size = Pt(10.5)  # 设置页眉字体大小
    # ph.bold = True  # 页脚字体加粗
    pfooter.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER  # 页眉对齐方式设为居中



# 在doc结尾处调用进行设置
# 通过sections（节）进行设置时，节对应文档中的每一页
# 每个节在没输入内容之前是不存在的,因此在最后才对每个节逐一进行设置
def setsectionformat(doc):
    for sec in doc.sections:
        # 设置页面边距(左上25毫米，右下15毫米)
        sec.top_margin = Cm(2.5)
        sec.left_margin = Cm(2.5)
        sec.right_margin = Cm(1.5)
        sec.bottom_margin = Cm(1.5)
        # 设置纸张大小(A4)
        sec.page_height = Mm(297)
        sec.page_width = Mm(210)
        # 设置页眉页脚距离
        sec.header_distance = Cm(1.5)
        sec.footer_distance = Cm(0.2)


# 获取测试用例执行集
csylzxj = xlrd2.open_workbook('./date/CESHIYONGLIZHIXINGJI.xlsx')
# czxtcssmb = xlrd.open_workbook('./file/ceshi')
document = Document()
# 调用函数设置全局文本样式变量
docxinitial(document)

# 设置全局变量
# 获取所有的sheet数量
sheet_sum = len(csylzxj.sheet_names())
# 获取所有的sheet的名字
sheet_names = csylzxj.sheet_names()
# 获取每个sheet中的用例个数，并计算总数，还有比例
sheet_yl_sum_value = []
# 获取每个sheet对应的行数
sheet_yl_sum = {}
for i in range(1, sheet_sum):
    sheet_yl_sum[csylzxj.sheet_by_index(i).name] = csylzxj.sheet_by_index(i).nrows
    sheet_yl_sum_value.append(csylzxj.sheet_by_index(i).nrows - 2)

# 测试过程中所有的测试类型数量，文档审查、安装性测试、功能测试、、、
cslx_sum = sheet_sum - 1

# 2、写入内容
# 标题
document.add_heading("1 范围", level=1)
document.add_heading("1.1 标识", level=2)
# 添加有序列表
document.add_paragraph(
    '本文档已批准的标识号：', style='List Number'
)
document.add_paragraph(
    '文档标题：', style='List Number'
)
document.add_paragraph(
    '本文档适用的被测软件名称和版本：', style='List Number'
)
document.add_paragraph(
    '术语和缩略语：\n说明：本文档中的所有术语和缩略语在此解释，未出现的不需在此解释。', style='List Number'
)
document.add_heading("1.2 文档概述", level=2)
p_1_2 = document.add_paragraph('''
    本文档主要描述XXX操作系统测试设计与实现工作的结果，是测试执行的依据。
    正文内容包括测试执行进度和测试准备过程，测试项的层次化分解，测试用例的设计方法及实现，以及测试用例对测试需求的追溯。
附录部分包括附录A软件文档审查单、附录B需求追溯（测试设计）表和附录C测试用例集
''')
document.add_heading("1.3 被测软件概述", level=2)
document.add_heading("1.3.1 系统概述", level=3)
p_1_3_1 = document.add_paragraph('概述本文档所适用的操作系统的用途及组成，给出其用途，应使用系统结构图。')

document.add_heading("1.3.2 主要功能", level=3)
p_1_3_2 = document.add_paragraph('概述被测操作系统的功能、使命任务、结构组成、信息流程。')

document.add_heading("1.3.3 主要技术指标", level=3)
p_1_3_3 = document.add_paragraph('''逐一描述被测软件的主要技术指标。
a) 技术指标1
b) 技术指标2
c) 技术指标3
……
''')

document.add_heading("1.3.4 主要接口", level=3)
p_1_3_4 = document.add_paragraph('描述接口情况，应包含应用编程接口和硬件编程接口。')

document.add_heading("1.3.5 被测软件总体情况", level=3)
p_1_3_5 = document.add_paragraph('被测软件的总体情况见表1-1')
t_1_3_5 = document.add_paragraph('表1-1 被测软件总体情况')
t_1_3_5.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
# 添加一个5行3列的表格，样式是网格实线
table_1_1 = document.add_table(7, 2, style="Table Grid")
table_1_1.cell(0, 0).text = "软件产品名称"
table_1_1.cell(1, 0).text = "软件版本（首轮）"
table_1_1.cell(2, 0).text = "内核版本（首轮）"
table_1_1.cell(3, 0).text = "软件版本（回归）"
table_1_1.cell(4, 0).text = "内核版本（回归）"
table_1_1.cell(5, 0).text = "运行平台"
table_1_1.cell(6, 0).text = "承研单位"
table_1_1.cell(0, 1).text = "软件产品名称"
table_1_1.cell(1, 1).text = "软件版本（首轮）"
table_1_1.cell(2, 1).text = "内核版本（首轮）"
table_1_1.cell(3, 1).text = "软件版本（回归）"
table_1_1.cell(4, 1).text = "内核版本（回归）"
table_1_1.cell(5, 1).text = "运行平台"
table_1_1.cell(6, 1).text = "承研单位"
table_1_1.style.font.size = Pt(15)  # 字体大小15磅
# table1_1.style.font.color.rgb = RGBColor.from_string("6495ED")  # 字体颜色
table_1_1.style.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER  # 居中对齐

document.add_heading("2 依据和应用文件", level=1)
p_2 = document.add_paragraph('按表2-1的格式列出本测试说明所依据和引用的文件。')
t_2 = document.add_paragraph('表2-1 依据和引用文件')
t_2.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
table_2_1_nrow = 5
table_2_1 = document.add_table(table_2_1_nrow, 5, style="Table Grid")
table_2_1.cell(0,0).text = '标识'
table_2_1.cell(0,1).text = '文档名称'
table_2_1.cell(0,2).text = '版本'
table_2_1.cell(0,3).text = '发布日期'
table_2_1.cell(0,4).text = '颁布/来源单位'


document.add_heading("3 测试进度与准备", level=1)
document.add_heading("3.1 测试进度", level=2)
p_3_1 = document.add_paragraph('提供与测试准备、执行、数据整理和分析有关活动的进度。\n    本次动态测试的进度安排见表3-1。')
t_3_1 = document.add_paragraph('表3-1 动态测试进度安排')
t_3_1.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
table_3_1_nrow = 5
table_3_1 = document.add_table(table_3_1_nrow, 4, style="Table Grid")
table_3_1.cell(0,0).text = '测试活动'
table_3_1.cell(0,1).text = '活动说明'
table_3_1.cell(0,2).text = '周期（天）'
table_3_1.cell(0,3).text = '开始时间'

document.add_heading("3.2 测试准备", level=2)
p_3_2 = document.add_paragraph('分小节描述测试的准备和配置情况。')

document.add_heading("3.2.1 硬件准备", level=3)
p_3_2_1 = document.add_paragraph('''描述测试所需硬件的准备过程，也可引用已发布的软件文档，它应提供以下信息：
a.	所使用的特定硬件，并标识其名称和编号；
b.	连接硬件所需的所有开关设置和电缆，并标识其名称和位置；
c.	绘制硬件、联接控制和数据路径的一个或多个原理图；
d.	组装硬件的详细指令。
搭建测试平台所需的硬件设施的主要构成。测试平台的系统结构用图表示。
    测试所需硬件设施的名称和属性见表3-2。
''')
t_3_2_1 = document.add_paragraph('表3-2 硬件和固件项')
t_3_2_1.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
table_3_2_nrow = 5
table_3_2 = document.add_table(table_3_2_nrow, 6, style="Table Grid")
table_3_2.cell(0,0).text = '序号'
table_3_2.cell(0,1).text = '硬件和固件项名称'
table_3_2.cell(0,2).text = '规格和标识'
table_3_2.cell(0,3).text = '数量'
table_3_2.cell(0,4).text = '用途'
table_3_2.cell(0,5).text = '所属权'


document.add_heading("3.2.2 软件准备", level=3)
p_3_2_2 = document.add_paragraph('''描述为测试而准备计算机软件配置项和保障软件所需的过程及相关信息。
    测试环境所需的软件名称和属性见表3-3。
''')
t_3_2_2 = document.add_paragraph('表3-2 软件项')
t_3_2_2.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
table_3_3_nrow = 5
table_3_3 = document.add_table(table_3_2_nrow, 5, style="Table Grid")
table_3_3.cell(0,0).text = '序号'
table_3_3.cell(0,1).text = '软件项名称'
table_3_3.cell(0,2).text = '版本号'
table_3_3.cell(0,3).text = '用途'
table_3_2.cell(0,4).text = '提供方'


document.add_heading("3.2.3 其他测试准备", level=3)
p_3_2_3 = document.add_paragraph('''描述完成测试所需的任何其它准备工作。
在动态测试工作开展之前，还需完成其它测试准备如下：
a.	开发测试的脚本以及测试支持软件等；
b.	测试数据准备；
c.	测试数据生成软件；
d.	测试结果数据处理软件。
''')

document.add_heading("4 测试说明", level=1)
p_4 = document.add_paragraph('''本次测试设计共设计{}个测试用例，其标识、需求追溯、预置条件、测试步骤、预期结果等详细情况见附录C测试用例集。（在委托方许可的情况下可以将测试用例集单独形成《XX软件测试用例设计集》文档，作为本文档的引用文档。但必须是同型号同批任务按相同方式处理）。以下内容是根据测试类型对测试用例进行分类的简要介绍。测试用例的分布情况的统计信息见表4-1。
'''.format(sum(sheet_yl_sum_value)))
# 首行缩进两个字
# p_4.paragraph_format.first_line_indent = Pt(20) * 1

t_4_x = document.add_paragraph('表4-1 测试用例分布一览表')
t_4_x.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
table_4_1 = document.add_table(sheet_sum, 3, style="Table Grid")
table_4_1.cell(0, 0).text = '测试类型'
table_4_1.cell(0, 1).text = '测试用例个数'
table_4_1.cell(0, 2).text = '所占比例'
# 获取测试类型
for i in range(1, sheet_sum):
    table_4_1.cell(i, 0).text = str(sheet_names[i])
# 获取每个测试sheet中的测试用例数
for i in range(1, sheet_sum):
    table_4_1.cell(i, 1).text = str(sheet_yl_sum_value[i - 1])
# 获取各个sheet中的用例数在从的用例数中所占的比例
for i in range(1, sheet_sum):
    table_4_1.cell(i, 2).text = str('{:.2%}'.format(sheet_yl_sum_value[i - 1] / sum(sheet_yl_sum_value)))

document.add_heading("4.1 文档审查", level=2)
p_4_1 = document.add_paragraph('''
    简述测试大纲中文档审查要求。
    使用软件文档审查单（见附录A）记录文档审查工作，根据检查单对操作系统文档进行审查。
''')

# 用 cslx_num 来标识序列号4.x
cslx_num = 2
for i in range(1, cslx_sum):
    document.add_heading("4.{} {}".format(cslx_num, csylzxj.sheet_by_index(i).name), level=2)
    p_4_x = document.add_paragraph(
        '简述被测操作系统的{}要求，描述测试的设计思路。\n    测试用例说明见表4-{}'.format(csylzxj.sheet_by_index(i).name, cslx_num))
    # 首行缩进两个字
    p_4_x.paragraph_format.first_line_indent = Pt(20) * 1
    t_4_x = document.add_paragraph('表4-%d 安装性测试用例一览表' % cslx_num)
    t_4_x.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    xxcsyl = csylzxj.sheet_by_name('{}'.format(csylzxj.sheet_by_index(i).name))
    table_4_x = document.add_table(xxcsyl.nrows - 2, 2, style="Table Grid")
    table_4_x.cell(0, 0).text = '测试用例标识'
    table_4_x.cell(0, 1).text = '测试用例说明'
    for i in range(1, xxcsyl.nrows - 2):
        table_4_x.cell(i, 0).text = xxcsyl.cell_value(i + 1, 4)
        table_4_x.cell(i, 1).text = xxcsyl.cell_value(i + 1, 5)
    cslx_num += 1

document.add_heading("5 回归测试", level=1)
# 回归测试设计一览表
t_5 = document.add_paragraph('表5-1 回归测试设计一览表')
t_5.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
table_hggncsyl = document.add_table(1, 6, style="Table Grid")
table_hggncsyl.cell(0, 0).text = '序号'
table_hggncsyl.cell(0, 1).text = '测试项标识'
table_hggncsyl.cell(0, 2).text = '测试项'
table_hggncsyl.cell(0, 3).text = '测试用例标识'
table_hggncsyl.cell(0, 4).text = '测试用例说明'
table_hggncsyl.cell(0, 5).text = '状态'
# 设置回归测试用例的序号
num_hgcs = 1
# 遍历所有sheet找到问题标识列中不为空的数据，并添加到新的sheet中
for j in range(1, sheet_sum):
    sheet_x_5 = csylzxj.sheet_by_index(j)
    last_col3 = sheet_x_5.ncols
    for i in range(1, sheet_x_5.nrows - 1):
        if sheet_x_5.cell_value(i + 1, last_col3 - 1) != '':
            # 遍历所有sheet中的所有问题标识列，如果问题标识列不为空，那么就在回归测试表格末尾新添加一行
            new_add_row = table_hggncsyl.add_row().cells
            # 序号
            new_add_row[0].text = str(num_hgcs)
            # 测试项标识
            new_add_row[1].text = sheet_x_5.cell_value(i + 1, 0)
            # 测试项
            new_add_row[2].text = sheet_x_5.cell_value(i + 1, 1)
            # 测试用例标识
            new_add_row[3].text = sheet_x_5.cell_value(i + 1, 4)
            # 测试用例说明
            new_add_row[4].text = sheet_x_5.cell_value(i + 1, 5)
            # 状态
            new_add_row[5].text = '重用'
            num_hgcs += 1
# 在t_5前添加段落描述
p_5 = t_5.insert_paragraph_before('''
    描述回归测试阶段测评需求的变化或软件更改的情况，分析变化和更改对被测件造成的影响。
    根据影响域分析，本次回归测试的动态测试包含测试项{}个，其中新增测试项XX个；设计新测试用例XX个，重用测试用例XX个。全部的测试用例的详细信息见附录C测试用例集。\n    本轮回归测试的测试设计情况见表5-1。'''.format(table_hggncsyl.columns[0].cells[-1].text))

document.add_heading("6 需求追溯性", level=1)
document.add_heading("附录A 软件文档审查单", level=1)
document.add_heading("附录B 需求追溯（测试设计）表", level=1)
table_xqzsb_nrow = sum(sheet_yl_sum_value) + 2
table_xqzsb = document.add_table(table_xqzsb_nrow, 9, style="Table Grid")
table_xqzsb.cell(0, 0).text = '《xxx操作系统测试大纲》'
table_xqzsb.cell(0, 4).text = '测试设计'
# 合并单元格
table_xqzsb.cell(0, 3).merge(table_xqzsb.cell(0, 0))
table_xqzsb.cell(0, 8).merge(table_xqzsb.cell(0, 4))

table_xqzsb.cell(1, 0).text = '序号'
table_xqzsb.cell(1, 1).text = '需求或设计文档章节号'
table_xqzsb.cell(1, 2).text = '标识'
table_xqzsb.cell(1, 3).text = '状态'
table_xqzsb.cell(1, 4).text = '序号'
table_xqzsb.cell(1, 5).text = '测试用例名称'
table_xqzsb.cell(1, 6).text = '测试用例标识'
table_xqzsb.cell(1, 7).text = '测试类型'
table_xqzsb.cell(1, 8).text = '状态'
# 设置需求追溯的序号
num_xqzsb = 2
# 遍历所有sheet找到问题标识列中不为空的数据，并添加到新的sheet中
for j in range(1, sheet_sum):
    sheet_x_B = csylzxj.sheet_by_index(j)
    last_col3 = sheet_x_B.ncols
    for i in range(1, sheet_x_B.nrows - 1):
        # 序号
        table_xqzsb.cell(num_xqzsb, 0).text = str(num_xqzsb - 1)
        # 需求或设计文档章节号
        table_xqzsb.cell(num_xqzsb, 1).text = sheet_x_B.cell_value(i + 1, 0)
        # 标识
        table_xqzsb.cell(num_xqzsb, 2).text = sheet_x_B.cell_value(i + 1, 1)
        # 状态
        table_xqzsb.cell(num_xqzsb, 3).text = '原始'
        # 序号
        table_xqzsb.cell(num_xqzsb, 4).text = str(num_xqzsb - 1)
        # 测试用例名称
        table_xqzsb.cell(num_xqzsb, 5).text = sheet_x_B.cell_value(i + 1, 3)
        # 测试用例标识
        table_xqzsb.cell(num_xqzsb, 6).text = sheet_x_B.cell_value(i + 1, 4)
        # 测试类型
        table_xqzsb.cell(num_xqzsb, 7).text = sheet_x_B.name[:-2]
        # 状态
        table_xqzsb.cell(num_xqzsb, 8).text = '原始'
        num_xqzsb += 1

document.add_heading("附录C 测试用例集", level=1)
setsectionformat(document)
# 3、保存文件
document.save('./date/caozuoxitongceshishuoming.docx')
```
