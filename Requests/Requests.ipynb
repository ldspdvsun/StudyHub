## 发送请求

```python
from urllib.parse import urljoin

import requests
import json
```

```python
base = "http://httpbin.org"
baseGithub = "https://api.github.com"
```

### get

```python
url = urljoin(baseGithub, "/events")
r = requests.get(url)
print(r.text)
```

### post

```python
url = urljoin(base, "/post")
data = {"key": "value"}
r = requests.post(url, data)
print(r.text)
```

### put

```python
url = urljoin(base, "/put")
data = {"key": "value"}
r = requests.put(url, data=data)
print(r.text)
```

```python
url = urljoin(base, "/delete")
r = requests.delete(url)
print(r.text)
```

### head

```python
url = urljoin(base, "/put")
r = requests.head(url)
print(r.text)
```

### options

```python
url = urljoin(base, "/get")
r = requests.options(url)
print(r.text)
```

```python
url = urljoin(base, "/get")
r = requests.options(url)
print(r.encoding)
```

## 定制请求头

定制header的优先级低于某些特定的信息源，例如：
+ 如果在 .netrc 中设置了用户认证信息，使用 headers= 设置的授权就不会生效。而如果设置了 auth= 参数，“.netrc”的设置就无效了
+ 如果被重定向到别的主机，授权header就会被删除
+ 代理授权header会被URL中提供的代理身份覆盖掉
+ 在我们能判断内容长度的情况下，header和Content-Length会被改写

Requests不会基于定制header的具体情况改变自己的行为，只不过在最后的请求中，所有的header信息都会被传递进去

注意：所有的header值必须是string、bytestring或者unicode。尽管传递unicode header也是允许的，但不建议这样做。

```python
url = urljoin(baseGithub,'/some/endpoint')
headers = {'user-agent':'my-app/0.0.1'}
r = requests.get(url,headers=headers)
print(r.url)
print(r.text)
print(r.headers)
```

## 更加复杂的POST请求

想要发送一些编码为表单形式的数据--像一个HTML表单。要实现这个，只需要简单的传递一个字典给data参数，数据字典在发出请求是会自动编码为表单形式

```python
url = urljoin(base,'/post')
payload = {'key1':'value1','key2':'value2'}
r = requests.post(url,data = payload)
print(r.url)
print(r.text)
```

为data参数传入一个元组列表。在表单中多个元素使用同一key的时候，这种方式尤其有效

```python
url = urljoin(base,'/post')
payload = (('key1','value1'),('key1','value2'))
r = requests.post(url,data=payload)
print(r.text)
```

很多时候想要发送的数据并非表单形式的。如果传递一个string而不是一个dict，那么数据会被直接发布出去

```python
# url=urljoin(baseGithub,'/some/endpoint')
url = urljoin(base,'post')
payload={'some':'data'}
r = requests.post(url,data = json.dumps(payload))
print(r.text)
```

## POST 一个多部分编码（Multipart-Encoded）的文件

Requests使得上传多部分编码文件变得简单

```python
url = urljoin(base,'/post')
files = {'file':open('a.xlsx','rb')}

r=requests.post(url,files = files)
print(r.text)
```

显示地设置文件名，文件类型和请求头

```python
url = urljoin(base,'/post')
files={'file':('a.xlsx',open('a.xlsx','rb'),'application/vnd.ms-excel',{'Expires':0})}

r=requests.post(url,files = files)
print(r.text)
```

也可以发送作为文件来接收的字符串

```python
url = urljoin(base,'/post')
files = {'file':('a.xlsx','some,data,to,send\nanother,row,to,send\n')}

r = requests.post(url,files = files)
print(r.text)
```

### 响应状态码

检测响应状态码

```python
url = urljoin(base,'/get')
r = requests.get(url)
print(r.status_code)
print(r.status_code == requests.codes.ok)
```

如果发送一个错误请求（一个4XX客户端错误，一个5XX服务器响应错误），可以通过Response.raise_for_status()来抛出异常：

```python
badUrl = urljoin(base,'/status/404')
r = requests.get(badUrl)
print(r.status_code)
print(r.raise_for_status())
```

当 r 的 status_code 是200，调用 raise_for_status() 时，得到的是 None

```python
url = urljoin(base,'/get')
r = requests.get(url)
print(r.status_code)
print(r.raise_for_status())
```

## Cookie

如果某个响应中包含一些cookie，可以快速访问它们，要想发送cookies到服务器，可以使用cookies参数

```python
url = urljoin(base,'/cookies')
cookies = dict(cookies_are='working')

r= requests.get(url,cookies=cookies)
print(r.text)
```

Cookie的返回对象为RequestCookieJar，它的行为和字典类似，但接口更为完整，适合跨域名跨路径使用。可以把CookieJar传到Requests中：

```python
jar = requests.cookies.RequestsCookieJar()
jar.set('tasty_cookie','yum',domain='httpbin.org',path='/cookies')
jar.set('gross_cookie','blech',domain='httpbin.org',path='/elsewhere')

url=urljoin(base,'/cookies')
r=requests.get(url,cookies=jar)
print(r.text)
print('======================')
url2=urljoin(base,'/elsewhere')
r2=requests.get(url2,cookies=jar)
print(r2.text)
```

## 重定向与请求历史

默认情况下，除了HEAD，Requests会自动处理所有重定向。可以使用响应对象的history方法来追踪重定向

Response.history 是一个 Response 对象的列表，为了完成请求而创建了这些对象。这个对象列表按照从最老到最近的请求进行排序

```python
r = requests.get(baseGithub)
print(r.url)
print(r.status_code)
print(r.history)
```

如果使用的是GET、OPTIONS、POST、PUT、PATCH或者DELETE，可以通过allow_redirects参数禁用重定向处理

```python
url = 'http://github.com'
r = requests.get(url,allow_redirects=False)
print(r.status_code)
print(r.history)
```

如果使用了HEAD，也可以启动重定向

```python
r = requests.get(baseGithub,allow_redirects = True)
print(r.url)
print(r.history)
```

## 超时

可以高速requests在经过以timeout参数设定的秒数时间之后停止等待响应。基本上所有的生产代码都应该使用这一参数，如果不使用，程序可能会永远失去响应

```python
r = requests.get(base,timeout = 0.0001)
print(r.text)
```

注意：
timout仅对连接过程有效，与响应体的下载无关。timeout并不是整个下载响应的时间限制，而是如果服务器在timeout秒内没有应答，将会引发一个异常（更精确地说，是在timeout秒内没有从基础套接字上收到任何字节的数据时）If not timeout is specified explicitly, requests do not time out

## 错误与异常

遇到网络问题（如：DNS查询失败，拒绝连接等）时，Requests会抛出一个ConnectionError异常。

如果HTTP请求返回了不成功的状态码，Response.raise_for_status()会抛出一个HTTPError异常。

若请求超时，则抛出一个Timeout异常。

若请求超过了设定的最大重定向次数，则会抛出一个TooManyRedirects异常。

所有Requests显式抛出的异常都继承自requests.exception.RequestException

## 会话对象

会话对象可以跨请求保持某些参数。会在同一个session实例发出的所有请求之间保持Cookie，期间使用urllib3的connection pooling功能。如果向同一台主机发送多个请求，底层的TCP连接将会重用，从而带来显著的性能提升。

```python
s = requests.Session()
url = urljoin(base,'/cookies/set/sessioncookie/123456789')
s.get(url)

url1 = urljoin(base,'/cookies')
r = s.get(url1)

print(r.text)
```
