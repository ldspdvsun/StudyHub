# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

>执行%matplotlib inline后，不用再执行plt.show()，执行完plt.plot()就可以显示图片

```python
%matplotlib inline
```

## 折线图

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.xlabel('xlabel',fontsize=16)
plt.ylabel('ylabel',fontsize=16)
```

### 参数选择

#### 线条样式

>linestyle='-.'

字符|类型|字符|类型
---|---|---|---
`'-'`|实线|`'--'`|虚线
`'-.'`|虚点线|`':'`|点线
`'-'`|点|`','`|像素点
`'o'`|圆点|`'v`|下三角点
`'~'`|上三角点|`'<'`|左三角点
`'>'`|右三角点|`'1'`|下三叉点
`'2'`|上三叉点|`'3'`|左三叉点
`'4'`|右三叉点|`'s'`|正方点
`'p'`|五角点|`'*'`|星形点
`'h'`|六边形点1|`'H'`|六边形点2
`'+'`|加号点|`'X'`|乘号点
`'D'`|实心菱形点|`'d'`|瘦菱形点
`'_'`|横线点||

#### 线条颜色

>color='r'

字符|颜色
--|--
''b''|蓝色，blue
''g''|绿色，green
''r''|红色，red
''c''|青色，cyan
''m''|品红色，magenta
''y''|黄色，yellow
''k''|黑色，black
''w''|白色，white


```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'-.',color='r')
plt.xlabel('xlabel',fontsize=16)
plt.ylabel('ylabel',fontsize=16)
```

#### 关键点
>marker='o'(参考线条样式)

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'-.',marker='*')
```

#### 绘制多个折线

```python
x_numpy = np.arange(0,10,0.5)
plt.plot(x_numpy,x_numpy,'r-.<')
plt.plot(x_numpy,x_numpy**2,'g-.o')
plt.plot(x_numpy,x_numpy**3,'b-.*')
```

```python
x_numpy = np.arange(0,10,0.5)
plt.plot(x_numpy,x_numpy,'r-.<',
        x_numpy,x_numpy**2,'g-.o',
        x_numpy,x_numpy**3,'b-.*')
```

#### 线条宽度
>linewidth=2.5

```python
x=np.linspace(-10,10)
y=np.sin(x)
plt.plot(x,y,linewidth=2.5)
```

#### 标记点属性
>设置标记点样式：marker='o'

>设置标记点颜色：markerfacecolor='r'

>设置标记点大小：markersize=10)

```python
plt.plot(x,y,color='g',linestyle='-.',marker='o',markerfacecolor='r',markersize=10)
```

#### 设置透明程度
>alpha=0.5

```python
line=plt.plot(x,y)
plt.setp(line,color='r',linewidth=4.0,alpha=0.5)
```

#### 给图加上文字


```python
line=plt.plot(x,y)
plt.setp(line,color='m',linewidth=4.0,alpha=0.5,linestyle='-.',marker='o',markerfacecolor='k',markersize=10)
plt.xlabel('xlabel',fontsize=16)
plt.ylabel('ylabel',fontsize=16)
plt.title('x&y')
plt.text(0,0,'0,0-Text')
```

### 子图
>plt.subplot(211)

211不是数字二百一十一，而是分别代表**行，列，第几个**，所以

211 代表将要绘制的2行1列里的第1个子图

212 代表将要绘制的2行1列里的第2个子图

```python
plt.subplot(211)
plt.plot(x,y,color='r')
plt.subplot(212)
plt.plot(x,y,color='b')
```

```python
plt.subplot(121)
plt.plot(x,y,color='r')
plt.subplot(122)
plt.plot(x,y,color='b')
```
