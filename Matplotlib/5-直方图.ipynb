```python
import random

import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.normal(0, 20, 1000)
bins = np.arange(-100, 100, 5)
```

# 直方图

```python
plg, ax = plt.subplots()
ax.hist(data, bins=bins)
ax.set_xlim(min(data), max(data))
```

```python
data1 = [random.gauss(15, 10) for i in range(500)]
data2 = [random.gauss(5, 10) for i in range(500)]
plg, ax = plt.subplots()
ax.hist(data1, bins=bins, label="class1", alpha=0.3)
ax.hist(data2, bins=bins, label="class2", alpha=0.8)
ax.set_xlim(min(min(data1), min(data2)), max(max(data1), max(data2)))
ax.legend(loc="upper left")
```
