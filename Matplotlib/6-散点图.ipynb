```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# 散点图

```python
mu_vecl = np.array([0,0])
cov_matl = np.array([[2,0],[0,2]])

x1_samples = np.random.multivariate_normal(mu_vecl,cov_matl,100)
x2_samples = np.random.multivariate_normal(mu_vecl+0.2,cov_matl+0.2,100)
x3_samples = np.random.multivariate_normal(mu_vecl+0.4,cov_matl+0.4,100)

plg,ax = plt.subplots()
ax.scatter(x1_samples[:,0],x1_samples[:,1],marker='x',color='blue',alpha=0.5,label='x1')
ax.scatter(x2_samples[:,0],x2_samples[:,1],marker='o',color='red',alpha=0.5,label='x2')
ax.scatter(x3_samples[:,0],x3_samples[:,1],marker='*',color='green',alpha=0.5,label='x3')

ax.legend()
```

```python
x_coords=[random.randint(0,10) for i in range(10)]
y_coords=[random.randint(0,10) for i in range(10)]

plg,ax=plt.subplots()
ax.scatter(x_coords,y_coords,marker="*",s=90)

for x,y in zip(x_coords,y_coords):
    ax.annotate('(%s,%s)'%(x,y),xy=(x,y),xytext=(0,-15),textcoords='offset points',ha='center')
```

```python
mu_vecl = np.array([0,0])
cov_matl = np.array([[1,0],[0,1]])
X = np.random.multivariate_normal(mu_vecl,cov_matl,500)

R=X**2
R_sum=R.sum(axis=1)

plg,ax=plt.subplots()
ax.scatter(X[:,0],X[:,1],color='grey',marker='o',s=20*R_sum)
```
