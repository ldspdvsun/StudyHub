```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.use("nbagg")
%matplotlib inline
%config InlineBackend.figure_format = "svg"
```

## 条形图方向

* 纵向 ax.bar()
* 横向 ax.barh()

```python
np.random.seed(0)
x = np.arange(5)
y = np.random.randint(-5, 5, 5)

fig, axes = plt.subplots(1, 2)
# 竖向 axes[0].bar()
v_bars = axes[0].bar(x, y, color="m")
# 横向 axes[1].barh()
h_bars = axes[1].barh(x, y, color="b")

# 添加横线
# axes[0].axhline添加横向线条
# axes[1].axvline添加竖向线条
# axes[子图索引].axhline(横线位置,color='gray',linewidth=2)
axes[0].axhline(0, color="gray", linewidth=2)
axes[1].axvline(0, color="gray", linewidth=3)
plt.show()
```

## 条形图值判断

```python
fig, axes = plt.subplots()
v_bars = axes.bar(x, y, color="lightblue")
for bar, height in zip(v_bars, y):
    if height < 0:
        bar.set(edgecolor="darkred", color="green", linewidth=3)
```

## 条形图填充

```python
x = np.random.randn(100).cumsum()
y = np.linspace(0, 10, 100)

fig, ax = plt.subplots()
ax.fill_between(x, y, color="lightblue")
```

```python
x = np.linspace(0, 10, 200)
y1 = 2 * x + 1
y2 = 3 * x + 1.2
y_mean = 0.5 * x * np.cos(2 * x) + 2.5 * x + 1.1

fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, color="red", alpha=0.5)
ax.plot(x, y_mean, color="black")
```

## 条形图误差值

```python
# 条形图的值
mean_values = [1, 2, 3]
# 误差
variance = [0.2, 0.4, 0.5]
# 标签
bar_label = ["bar1", "bar2", "bar3"]

# 位置
x_pos = list(range(len(bar_label)))
# 高度
max_y = max(zip(mean_values, variance))

fig, ax = plt.subplots()
# 画图
ax.bar(x_pos, mean_values, yerr=variance, color="red", alpha=0.4)

# 限制与y轴
ax.set_ylim([0, (max_y[0] + max_y[1]) * 1.2])

ax.set_ylabel("variable y")
ax.set_xticks(x_pos, bar_label)
```

## 双条形图 X轴正负轴展示

```python
x1 = np.array([1, 2, 3])
x2 = np.array([1, 2, 3])

bar_labels = ["bar1", "bar2", "bar3"]
fig, ax = plt.subplots()
y_pos = np.arange(len(x1))
print(y_pos)

y_pos = [x for x in y_pos]
print(y_pos)

ax.barh(y_pos, x1, color="g", alpha=0.5)
ax.barh(y_pos, -x2, color="r", alpha=0.5)

# 设置x轴限制
ax.set_xlim(-max(x2) - 1, max(x1) + 1)
# 设置y轴限制
ax.set_ylim(-1, len(x1) + 0.5)
```

## 多条形图 以单个坐标轴为单位循环展示

```python
green_data = [1, 2, 3]
yellow_data = [3, 2, 1]
red_data = [3, 2, 3]

labels = ["gropp1", "group2", "group3"]
pos = list(range(len(green_data)))
print(pos)
width = 0.1
fig, ax = plt.subplots()

ax.bar(pos, green_data, width, alpha=0.5, color="green", label=labels[0])
ax.bar(
    [p + width for p in pos],
    yellow_data,
    width,
    alpha=0.5,
    color="yellow",
    label=labels[1],
)
ax.bar(
    [p + width * 2 for p in pos],
    red_data,
    width,
    alpha=0.5,
    color="red",
    label=labels[2],
)

ax.legend(labels, loc="upper center")
```

## 多条形图 以每个条形图组为单位展示

```python
pos = list(range(len(green_data)))
print(pos)
width = 0.5
fig, ax = plt.subplots()

ax.bar(pos, green_data, width, alpha=0.5, color="green", label=labels[0])
ax.bar(
    [p + len(pos) for p in pos],
    yellow_data,
    width,
    alpha=0.5,
    color="yellow",
    label=labels[1],
)
ax.bar(
    [p + 2 * len(pos) for p in pos],
    red_data,
    width,
    alpha=0.5,
    color="red",
    label=labels[2],
)

ax.legend(labels, loc="upper center")
```

### 更改坐标轴刻度值以及步长

### 更改坐标轴坐标值，替换为指定内容

```python
pos = list(range(1, len(green_data) + 1))
print(pos)
width = 0.5
fig, ax = plt.subplots()

ax.bar(pos, green_data, width, alpha=0.5, color="green", label=labels[0])
ax.bar(
    [p + len(pos) for p in pos],
    yellow_data,
    width,
    alpha=0.5,
    color="yellow",
    label=labels[1],
)
ax.bar(
    [p + 2 * len(pos) for p in pos],
    red_data,
    width,
    alpha=0.5,
    color="red",
    label=labels[2],
)

ax.legend(labels, loc="upper left")
ax.set_xlim(pos[0] - 1, 3 * len(pos) + 1)

x_lab = ["q", "w", "e", "r", "t", "y", "u", "i", "o"]
x_pos = np.arange(1, 10, 1)
print(x_lab)
print(x_pos)

ax.xaxis.set_ticks(x_pos, x_lab)
```

### 条形图值展示

```python
data = range(200, 225, 5)
bar_labels = ["a", "b", "c", "d", "e"]

fig, ax = plt.subplots()
y_pos = np.arange(len(data))
ax.yaxis.set_ticks(y_pos, bar_labels, fontsize=12)

bars = ax.barh(y_pos, data, alpha=0.5, color="g")
print(bars)

ax.vlines(min(data), -1, len(data) + 0.5, linestyle=":")

# 更改x轴长度，如果不加，条形图的值将会超出画布
ax.set_xlim(0, 270)

for b, d in zip(bars, data):
    print(b)
    ax.text(
        b.get_width() + b.get_width() * 0.05,
        b.get_y() + b.get_height() / 2,
        "{0:.2%}".format(d / min(data)),
    )
```
