# 风格设置

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.use('nbagg')
%matplotlib inline
```

```python
plt.style.available
```

```python
x=np.linspace(-10,10)
y=np.sin(x)
plt.plot(x,y)
```

## 风格选择
从plt.style.available获得的值中选择

```python
plt.style.use('seaborn-whitegrid')
plt.plot(x,y)
```

```python
plt.style.use('seaborn-notebook')
plt.plot(x,y)
```

```python
plt.xkcd()
plt.plot(x,y)
```
